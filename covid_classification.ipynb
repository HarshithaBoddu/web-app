```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['risk'] = le.fit_transform(data['risk'])
data['Appointment'] = le.fit_transform(data['Appointment'])
```

```python
data.drop(['Appointment'], axis = 1, inplace = True)
```

```python
x = data.iloc[:, 0:7].values
y = data.iloc[:, 7:8].values
```

```python
from sklearn.preprocessing import OneHotEncoder
oh = OneHotEncoder()
z=oh.fit_transform(y[:,0:1]).toarray()
```

```python
z
```

```python
y=np.delete(y,[0],axis=1) 
```

```python
y=np.concatenate((z,y),axis = 1)
```

```python
y.shape
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

LOGISTIC REGRESSION

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(random_state = 0)# random _state=0 is for consistency...to use the same set every time we run the program
classifier.fit(x_train,y_train)
```

```python
decisiontree = classifier.predict(x_test)
```

```python
decisiontree
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,decisiontree)
```
